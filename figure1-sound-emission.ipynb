```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib.pyplot import *

%matplotlib inline
```

```python
sns.set_theme(context='talk', style='white', font="Helvetica Neue", palette="tab10")
sns.set_style('ticks')
```

```python
df = pd.read_feather('vocalization_df.feather')

df_c2 = df[df.cohort=='c2']
df_c4 = df[df.cohort=='c4']
df_c5 = df[df.cohort=='c5']
```

```python
#sound events (from audio segmentation on all files for all families)
cohort2_sound_events = 2576616
cohort4_sound_events = 3074279
cohort5_sound_events = 4033840

figure(figsize=(3,6))
bar([0,1,2], np.array([cohort2_sound_events,
                       cohort4_sound_events,
                       cohort5_sound_events])/1e6, color=['#fff400', '#fff400', '#fff400'], 
    width=0.9)
ylabel('Total Sound Events (Millions)')
xticks([0,1,2], ('Family 1', 'Family 2', 'Family 3'), rotation=45);
twinx()
bar([0,1,2], np.array([cohort2_sound_events,
                       cohort4_sound_events,
                       cohort5_sound_events])/20/24/1000, color=['#fff400', '#fff400', '#fff400'],
    width=0.9, edgecolor = "black")
ylabel('Sound Events/Hour (Thousands)', labelpad=10)
yticks([2,4,6,8]);

# tight_layout()
# savefig('sound_events_bar.svg')
```

```python
#vocalization sound event proportion
vox_prop = np.array([len(df_c2), len(df_c4), len(df_c5)])/np.array([cohort2_sound_events,
                                                                    cohort4_sound_events,
                                                                    cohort5_sound_events])

figure(figsize=(3,6))
bar([0,0,1,1,2,2], [1, 1-vox_prop[0], 1, 1-vox_prop[1], 1, 1-vox_prop[2]], 
    color=['#17abeb', '#e91a23'], 
    width=0.9, edgecolor = "black")
xticks([0,1,2], ('Family 1', 'Family 2', 'Family 3'), rotation=45);
ylabel('Proportion sound events')
ylim(.7,1.01);
# savefig('vox_non_vox_bar.svg')
```

```python
#vocalizations
figure(figsize=(3,6))

bar([0,1,2], np.array([len(df_c2), len(df_c4), len(df_c5)])/1000, color=['#17abeb', '#17abeb', '#17abeb'], 
    width=0.9, edgecolor = "black")
ylabel('Total  # Vocalizations (Thousands)')
xticks([0,1,2], ('Family 1', 'Family 2', 'Family 3'), rotation=45);
twinx()
bar([0,1,2], np.array([len(df_c2), len(df_c4), len(df_c5)])/20/24, color=['#17abeb', '#17abeb', '#17abeb'], 
    width=0.9, edgecolor = "black")
ylabel('# Vocalizations/Hour', labelpad=10);
# savefig('vox_barsvg')
```
